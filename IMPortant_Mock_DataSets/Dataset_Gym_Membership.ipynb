### Dataset: Gym Membership Analytics

```python
import numpy as np

# Columns:
# [Member_ID, Age, Membership_Type,
#  Monthly_Fee, Months_Joined,
#  Trainer_Fee, Attendance_Days,
#  Rating]

a = np.array([
    [2001, 22, 1, 1500, 12, 500, 22, 4],
    [2002, 35, 2, 2500, 8, 1000, 18, 5],
    [2003, 28, 1, 1500, 6, 500, 15, 3],
    [2004, 45, 3, 3500, 10, 1500, 25, 5],
    [2005, 31, 2, 2500, 5, 1000, 12, 4],
    [2006, 26, 1, 1500, 9, 500, 20, 2],
    [2007, 39, 3, 3500, 7, 1500, 27, 5],
    [2008, 29, 2, 2500, 4, 1000, 10, 3],
    [2009, 50, 3, 3500, 12, 1500, 30, 4],
    [2010, 23, 1, 1500, 11, 500, 24, 5],
    [2011, 37, 2, 2500, 6, 1000, 14, 2],
    [2012, 42, 3, 3500, 8, 1500, 26, 5]
])
```

```python
# Q1. Total Membership Fee Collected
tmf=np.sum(a[:,3]*a[:,4])# Total membership fee
print(tmf)
```

```python
# Q2. Average Age Of Members

aa=np.mean(a[:,1]) # Average age
print(aa)

```

```python
# Q3. Maximum Attendance Days

mad=np.max(a[:,-2])# maximum attendance days
print(mad)
```

```python
# Q4. Count Members With Rating 5

mr=a[:,-1]==5#members rating
print(np.sum(mr))
```

```python
# Q5. Members Attending More Than 20 Days

ar=a[a[:,-2]>20]
print(ar)
```

```python
# Q6. Create Revenue Column

r=(a[:,3]+a[:,5])*a[:,4]#revenue
rc=np.column_stack((a,r))# revenue coolumn
print(rc)
print(rc.shape)
```

```python
# Q7. Member Generating Highest Revenue

r=(a[:,3]+a[:,5])*a[:,4]
mhr=np.argmax(r)# member highest revenue
print(a[mhr])
print(max(r))# maximum revenue
```

```python
# Q8. Total Gym Revenue

r=(a[:,3]+a[:,5])*a[:,4]
gr=np.sum(r)# gym revenue
print(gr)
```

```python
# Q9. Average Rating Per Membership Type

for cat in np.unique(a[:,2]):
    rows = a[a[:,2] == cat]
    avg_rating = np.mean(rows[:,-1])
    print(f"Membership Type {cat} -> {avg_rating}")
```

```python
# Q10. Age > Average Age AND Rating < Average Rating
c=(a[:,1]>np.mean(a[:,1])) &  (a[:,-1]<np.mean(a[:,-1]))
print(a[c])
```

```python
# Q11. Top 3 Members By Revenue

r=(a[:,3]+a[:,5])*a[:,4]

sr=np.argsort(r) # sorted revenue
print(a[sr[-1:-4:-1]])
print(np.sort(r)[-3:])


# print(a[np.argsort(r)[::-1][:3]]) another approach for top 3
```

```python
# Q12. Membership Type Generating Maximum Revenue

r=(a[:,3]+a[:,5])*a[:,4]
max_rev = 0
best_type = None

for cat in np.unique(a[:,2]):
    total_rev = np.sum(r[a[:,2] == cat])

    if total_rev > max_rev:
        max_rev = total_rev
        best_type = cat

print(f"Membership Type {best_type} -> {max_rev}")    
    
    
```

```python
# Q13. Members With Revenue Above Average Revenue
r=(a[:,3]+a[:,5])*a[:,4]
aar=r>np.mean(r)#above avg revenue 
print(a[aar])

avg=np.mean(r)# avg revenue
print(avg)
```

```python
#Q14. Create Performance Status Column
att = a[:,-2]

status = np.where(att >= 25, 3,
         np.where(att >= 15, 2, 1))

sc=np.column_stack((a,status))

print(status)
print(sc)
```

```python
# Q15. Correlation Matrix @@
corr = np.corrcoef(
    [a[:,1], a[:,3], a[:,6], a[:,7]]
)

print(np.round(corr, 2))
print(corr.shape)
```

```python
# Q16. Normalize @@

cols = a[:, [3,5,6]]
norm=(cols-np.min(cols,axis=0))/(np.max(cols,axis=0)-np.min(cols,axis=0))
print(np.round(norm,2))
```

```python
#Q17. Revenue Closest To Average Revenue
r=(a[:,3]+a[:,5])*a[:,4]

avg_r=np.mean(r)

rcar=np.argmin(np.abs(r-avg_r))
print(a[rcar])
print(r[rcar])

```

```python
#Q18. Revenue Contribution %
r=(a[:,3]+a[:,5])*a[:,4]

tr=np.sum(r)

for cat in np.unique(a[:,2]):
    category_rev = np.sum(r[a[:,2] == cat])
    
    contri=(category_rev/tr)*100
    print(f"Type {cat} -> {contri:.2f}%")
    
```

```python
# Q19. Attendance > Average Attendance AND Revenue > Average Revenue
r=(a[:,3]+a[:,5])*a[:,4]

q=(a[:,6]>np.mean(a[:,6])) & (r>np.mean(r))

print(a[q])
```

```python
r=(a[:,3]+a[:,5])*a[:,4]

s=(r*(a[:,-1]))/(a[:,-2]+1)



print(np.round(s,2))# score



bm=np.argmax(s)# Best Member
print(a[bm])
print(s[bm])



wm=np.argmin(s)# worst member
print(a[wm])
print(s[wm])



top5 = np.argsort(s)[-5:][::-1]

print(a[top5])

for i in top5:
    print(s[i])
```
