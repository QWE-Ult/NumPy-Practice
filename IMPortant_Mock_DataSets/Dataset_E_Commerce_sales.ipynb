```python
# Age > Average Age AND Rating < Average Rating

print(a[(a[:,1]>np.mean(a[:,1])) & (a[:,-2]<np.mean(a[:,-2]))])
```

```python
# Top 3 Orders By Final Revenue

r=a[:,3]*a[:,4]

d=r*(a[:,5]/100)

f=r-d


top3 = np.argsort(f)[-1:-4:-1]

print(f[top3])   # Top 3 revenues
print(a[top3])   # Corresponding rows

```

```python
# Q12. Category Generating Maximum Revenue  

r = a[:,3] * a[:,4]
d = r * (a[:,5] / 100)
f = r - d

max_category = None
max_revenue = 0

for cat in np.unique(a[:,2]):
    total_rev = np.sum(f[a[:,2] == cat])

    if total_rev > max_revenue:
        max_revenue = total_rev
        max_category = cat

print("Category:", max_category)
print("Revenue:", max_revenue) 
```

```python
# Customer Whose Revenue Is Closest To Average Revenue

r = a[:,3] * a[:,4]
d = r * (a[:,5] / 100)
f = r - d

print((a[np.argmin(np.abs(f - np.mean(f)))]))#order of closest to average
print(np.mean(f))#avereage 
print(f[np.argmin(np.abs(f - np.mean(f)))])# closest to average
```

```python
# Category-wise Revenue Percentage Contribution
r = a[:,3] * a[:,4]
d = r * (a[:,5] / 100)
f = r - d
tr=np.sum(f)# total revenue

for cat in np.unique(a[:,2]):
    cr = np.sum(f[a[:,2] == cat])# category revenue
    
    c=(cr/tr)*100
    print(cat,c)


```

```python
# Score
r = a[:,3] * a[:,4]
d = r * (a[:,5] / 100)
f = r - d

score = (f * a[:,6]) / (a[:,7] + 1)

best_idx = np.argmax(score)
worst_idx = np.argmin(score)
top5_idx = np.argsort(score)[-5:][::-1]

print("Best Order:\n", a[best_idx])
print("Best Score:\n", score[best_idx])

print("\nWorst Order:\n", a[worst_idx])
print("Worst Score:\n", score[worst_idx])

print("\nTop 5 Orders:\n", a[top5_idx])
print("\nTop 5 Scores:\n", score[top5_idx])
```
