```python
import numpy as np

# Data structure:
# [patient_id, BP, Sugar, Cholesterol, HeartRate]

a= np.array([
    [201, 120,  90, 180, 72],
    [202, 135, 110, 220, 85],
    [203, 118,  95, 175, 70],
    [204, 145, 130, 240, 92],
    [205, 125, 100, 190, 78],
    [206, 138, 115, 210, 88]
])

print("\n",a[::-1],"\n")

print("\n",a[:,:-5:-1],"\n")

print("\n",a[np.argmax(np.sum(a[:,1:],axis=1)),0],"\n")

print("\n",a[np.argsort(a[:,4]),0],"\n")

# Both the methods add Columns

print("\n", np.column_stack((a, np.sum(a[:,1:], axis=1))).shape, "\n")

print("\n", np.hstack((a, np.sum(a[:,1:], axis=1).reshape(-1, 1))).shape, "\n")


# For rows

print("\n", np.vstack((a, np.mean(a, axis=0))), "\n")      # Full array with average row

print("\n", np.round(np.mean(a, axis=0), 2), "\n")         # Only the average row


# Replace values

print("\n", np.where(a[:,1:] < 100, 100, a[:,1:]), "\n")

print("\n", np.where(a[:,1:] > 220, 220, a[:,1:]), "\n")


# Top 3 IDs based on total health metrics

print("\n", a[np.argsort(np.sum(a[:,1:], axis=1))[-3:], 0], "\n")
```
