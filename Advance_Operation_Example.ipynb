### Advance Operation with Business Example

```python
import numpy as np
```

```python
# Data strucutre: [restaurant_id,2021,2022,2023,2024]

sales_data= np.array([
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Chai Point
])

print("==== Zomato sales analysis ==== ")
print("\n Sales data shape", sales_data.shape)
print("\n Sample data for 1st 3 restaurant:\n",sales_data[:3, 1:5])
```

```python
# total sales Per year

total=sales_data[:,1]
print("Total sale of 2021 for Zomato :", np.sum(total))
#Per specific year


print("total of per years respectively",
      np.sum(sales_data[:,1:],axis=0))
# total of per years respectively(every year sales total)

print("total sale upto data",
      np.sum(sales_data[:,1:]))
#total sale upto data(Total sale upto date according to given data)




# Minimun sales per restaurant

# Every rest. min sale (how much min sale every rest. from all years  )
print("Every rest. min sale ",
      np.min(sales_data[:,1:],axis=1))


# Specific rest. min sale (ndividual rest. minimum sale )
print("Specific Rest.(Biryani) min sale",np.min(sales_data[0:1,1:],axis=1))
print("Specific Rest.(Burger Point) min sale",np.min(sales_data[3:4,1:],axis=1))


#Max sale per year (which rest was highest in every year)

print("Max sale of rest. per year",np.max(sales_data[:,1:],axis=0))



# Average sales per restaurant
print(np.mean(sales_data[:,1:],axis=1))


# Cumulative Sum

print(np.cumsum(sales_data[:,1:],axis=1))
```

### Trying Matplotlib 

```python
import matplotlib.pyplot as plt

sales_data= np.array([
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Chai Point
])
cumsum=np.cumsum(sales_data[:,1:],axis=1)
print(cumsum)

plt.figure(figsize=(10,6))
plt.plot(np.mean(cumsum,axis=0))
plt.title("Average Cumulative sales across all Restaurant")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```
