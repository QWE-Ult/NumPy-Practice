```python
print("\n",a[0,:],"\n") # Print the first row
print("\n",a[-1,:],"\n") # Print the last row
print("\n",a[0:5,:],"\n") # Print the first 5 rows
print("\n",a[-1:-11:-1,:],"\n") # Print the last 10 rows
print("\n",a[:,0],"\n") # Print the first column
print("\n",a[:,-1],"\n") # Print the last column
print("\n",a[:,0:3],"\n") # Print the 1st three columns
print("\n",a[10:21,2:6],"\n") # Print rows 10 to 20 and columns 2 to 5.
print("\n",a[0::10,:],"\n")#Print every 10th row of the dataset.
print("\n",a[::-1],"\n") # Reverse the entire dataset (reverse the order of rows).


```

##### Aggregation

```python
print("\n",np.mean(a[:,:]),"\n") # Find the overall mean of all 3,330,600 values in the NumPy array.

print("\n",np.mean(a[:,:],axis=0),"\n") # Find the mean of each column.

print("\n",np.sum(a[:,:],axis=0),"\n") # Find the sum of each column.

print("\n",np.min(a[:,:],axis=0),"\n") # Find the minimum value of each column.

print(np.max(a, axis=0)) # Find the maximum value of each column.

print(np.median(a, axis=0)) # Find the median (middle value) of each column.

print(np.std(a, axis=0)) # Find the standard deviation of each column.

print(np.var(a, axis=0)) # Find the variance of each column. 

print(np.argmax(a[:, -1])) # Find the index of the player with the highest tournament rating.

print(np.argmin(a[:, -1])) # Find the index of the player with the lowest tournament rating.

print(np.max(a[:, -1])) # Find the highest tournament rating.

print(np.min(a[:, -1])) # Find the lowest tournament rating.

print(a[a[:, 0] > 30]) #Find all players whose age is greater than 30 years.

print(np.unique(a[:, 0])) # Find all unique ages present in the dataset.

print(np.sort(a[:, -1])) # Sort the tournament ratings in ascending order.

print(np.argsort(a[:, -1])) # Find the indices that would sort the tournament ratings.
```
