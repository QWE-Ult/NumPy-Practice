### Array Reshaping

```python
arr=np.array([[1,2,3],
              [4,5,6]])
print("O\n",arr)

re=arr.reshape((3,2))# returs view (doesnt create copy(may affect original array))
print("R\n",re)

flat=re.flatten()
print("F\n",flat)
#returns copy(creates new array)
#changing in flatten array doesn't affect original array

ravel=re.ravel()
print("Ravel\n",ravel)
# returns view(original array), instead of copy whenever possible
#changing in ravel array may affect original array
#ravel() uses same memory
#Less memory used
#ravel() may sometimes create copy if needed internally.


# both flatten() and ravel() 
# convert multidimensional arrays into 1D arrays.

#Transpose

trans=re.T
print("T\n",trans)
```
