```python
import numpy as np

# Data structure:
# [store_id, 2021, 2022, 2023, 2024]

a= np.array([
    [1, 100000, 120000, 150000, 180000],
    [2,  90000, 110000, 130000, 170000],
    [3, 150000, 170000, 200000, 240000],
    [4, 130000, 145000, 175000, 210000],
    [5, 110000, 125000, 155000, 190000],
    [6, 160000, 190000, 220000, 260000]
])


print("\n",np.sum(a[:,1:],axis=1))#Q9. Find total sales of each store across all years.

print("\n",np.sum(a[:,1:],axis=0))#Q10. Find total sales for each year.

print("\n",np.mean(a[:,1:],axis=1))#Q11 Average sales of each store

print("\n",np.mean(a[:,1:],axis=0))#Q11 Average sales of each year

print("\n",np.max(a[:,:]))#Q13 Maximum sale value

print("\n",np.min(a[:,1:]))#Q14 Minimum sale value

print(f"\n store-ID={(np.argmax(a[:,4]))+1},","Store Index=",np.argmax(a[:,4]))#Q15. Find store with highest 2024 sales

print(f"\n store-ID={(np.argmin(a[:,1]))+1},","Store Index=",np.argmin(a[:,1]))#Q15. Find store with lowest 2021 sales

print("\n",a[a[:,4]>200000,0])#Q17 Stores whose 2024 sales > 200000

print("\n",a[a[:,3]>180000,0])#Q18. Print storeids where 2023 sales exceed 180000.
print("\n",a[a[:,3]>180000])#Q18. Print rows where 2023 sales exceed 180000.


print(np.sum(a[:,4]>200000))#Q19 Count stores having 2024 sales > 200000
```
